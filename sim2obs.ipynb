```python
import pandas as pd
import math
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
%matplotlib inline
import sys

# path to the scripts shared by Christine
tools_path = '../arepo-snap-util'
sys.path.append(tools_path)
from gadget_subfind import *

# path to the simulation data, change this to your path
sim_path = '/Users/RaindragonD/Dropbox/sims2obs'
```

```python
# get the simulation data as a dictionary, the keys are the acronyms above
def getSimData(select_halo):
    # Load data.  This call by default converts units to phyical units (so removes factors of a and h).
    path = sim_path+'/level4_MHD/halo_'+str(select_halo)+'/'
    sf = load_subfind(127,dir=path,hdf5=True)

    # Center subhalo positions on the host halo position
    spos_now = sf.data['spos']-sf.data['fpos'][0]
    # Compute distance from host
    sdist = (spos_now[:,0]**2 + spos_now[:,1]**2 + spos_now[:,2]**2)**0.5
    sf.data['sdist'] = sdist * 1e3
    sf.data['spos'] = sf.data['spos'] * 1e3
    # Compute subhalo velocity relative to host
    svel_rel = sf.data['svel']-sf.data['fvel'][0]
    sf.data['svel_rel'] = svel_rel
    # compute subhalo speed
    sf.data['sspeed'] = np.sqrt(np.sum(svel_rel**2, axis=1))
    # convert mass-related quantities to units of M_sun
    sf.data['smas'] = sf.data['smas'] * 1e10
    sf.data['smir'] = sf.data['smir'] * 1e10
    sf.data['smit'] = sf.data['smit'] * 1e10
    sf.data['smty'] = sf.data['smty'] * 1e10
    # extract some mass data (smty0=gas, smty4=star)
    sf.data['smty0'] = sf.data['smty'][:,0]
    sf.data['smty4'] = sf.data['smty'][:,4]
    # V-band magnitude
    sf.data['ssphv'] = sf.data['ssph'][:,2]
    return sf.data

# convert simulation quantity to its acronym if necessary
def checkSimVarName(name):
    if name in sim_var_acronyms.keys():
        name = sim_var_acronyms[name]
    return name

# plot relationships for both sim and obs
def plot2(sx, sy, ox, oy):
    sx_data = sim[checkSimVarName(sx)]
    sy_data = sim[checkSimVarName(sy)]
    ox_data, oy_data = obs[ox], obs[oy]
    
    plt.subplots(1,1)
    semilogy(sx_data, sy_data,'ro',alpha=0.5,label='sim')
#     xlabel(sim_var_units[sx])
#     ylabel(sim_var_units[sy])
#     title(sy + ' vs ' + sx)
    
    semilogy(ox_data, oy_data, 'bo',alpha=0.7,label='obs')
    legend(loc=0,numpoints=1)
    xlabel(obs_var_units[ox])
    ylabel(obs_var_units[oy])
    title(oy + ' vs ' + ox)
```

```python
def MHI_dist_graph(sim_num):
    sim = getSimData(sim_num)

    sx, sy, ox, oy = 'SubhaloDis', 'SubhaloMassTypeGas', 'min_dist_mw_or_m31', 'MHI_computed'

    sx_data = sim[checkSimVarName(sx)]
    sy_data = sim[checkSimVarName(sy)]
    ox_data, oy_data = obs[ox], obs[oy]

    sim_inds = sx_data <= 3000
    sx_data = sx_data[sim_inds]
    sy_data = sy_data[sim_inds]

    obs_inds = ox_data <= 3000
    ox_data = ox_data[obs_inds]
    oy_data = oy_data[obs_inds]

    fig, ax1 = plt.subplots(1,1)
    fig.set_size_inches(16.5, 10.5)

    ax1.semilogy(sx_data, sy_data,'ro',alpha=0.5,label='sim: with gas')

    detected_inds = obs['MHI_type']=='detected_mass'
    ax1.semilogy(ox_data[detected_inds], oy_data[detected_inds], 'bo',alpha=0.7,label='obs: detected')

    limit_inds = obs['MHI_type']!='detected_mass'
    ax1.semilogy(ox_data[limit_inds], oy_data[limit_inds], 'b+',alpha=0.7,label='obs: limit', markersize=10)

    ax1.legend(loc=2,numpoints=1)
    ax1.set_xlabel(obs_var_units[ox])
    ax1.set_ylabel(obs_var_units[oy])

    ax2 = ax1.twinx()
    no_gas_inds = sy_data == 0
    num_bins = 20
    n, bins, patches = ax2.hist(sx_data[no_gas_inds], num_bins, facecolor='red', alpha=0.5, label='sim: no gas')
    plt.ylim(0, 2000)
    bin_centers = 0.5*(bins[1:]+bins[:-1])
    ax2.plot(bin_centers,n, c='r')
    ax2.legend(loc=0,numpoints=1)
    ax2.set_ylabel('number')

    title('Mass of HI vs Distance to Host Galaxy (simulation %d)' % sim_num)
    plt.savefig("M_HI-vs-distance(%d)" % (sim_num))
```

```python
for i in range(1,31):
    MHI_dist_graph(i)
```

```python
# descision tree
from sklearn import tree
import graphviz

sim_num = 1
while sim_num <= 30:
    sim = getSimData(sim_num)
    X = np.array([sim[checkSimVarName(attribute)] for attribute in attributes]).T
    X.shape

    cutoff = 0
    y = sim[checkSimVarName('SubhaloMassTypeGas')] > cutoff
    sum(y)

    clf = tree.DecisionTreeClassifier(max_depth=3)
    clf = clf.fit(X, y)

    dot_data = tree.export_graphviz(clf, out_file=None, 
                          feature_names=attributes,  
                          class_names=['WithoutGas', 'WithGas'],  
                          filled=True, rounded=True,  
                          special_characters=True)  
    graph = graphviz.Source(dot_data) 
    graph.render('halo_'+str(sim_num))
    
    sim_num += 1
```
