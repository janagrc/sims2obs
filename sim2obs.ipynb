```python
import pandas as pd
import math
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
%matplotlib inline
import sys

# path to the scripts shared by Christine
tools_path = '../arepo-snap-util'
sys.path.append(tools_path)
from gadget_subfind import *

# path to the simulation data, change this to your path
sim_path = '/Users/RaindragonD/Dropbox/sims2obs'

obs = pd.read_csv('./dwarfs.csv',header=(0), comment='#')
```

```python
# get the simulation data as a dictionary, the keys are the acronyms above
def getSimData(select_halo):
    # Load data.  This call by default converts units to phyical units (so removes factors of a and h).
    path = sim_path+'/level4_MHD/halo_'+str(select_halo)+'/'
    sf = load_subfind(127,dir=path,hdf5=True)

    # Center subhalo positions on the host halo position
    spos_now = sf.data['spos']-sf.data['fpos'][0]
    # Compute distance from host
    sdist = (spos_now[:,0]**2 + spos_now[:,1]**2 + spos_now[:,2]**2)**0.5
    sf.data['sdist'] = sdist * 1e3
    sf.data['spos'] = sf.data['spos'] * 1e3
    # Compute subhalo velocity relative to host
    svel_rel = sf.data['svel']-sf.data['fvel'][0]
    sf.data['svel_rel'] = svel_rel
    # compute subhalo speed
    sf.data['sspeed'] = np.sqrt(np.sum(svel_rel**2, axis=1))
    # convert mass-related quantities to units of M_sun
    sf.data['smas'] = sf.data['smas'] * 1e10
    sf.data['smir'] = sf.data['smir'] * 1e10
    sf.data['smit'] = sf.data['smit'] * 1e10
    sf.data['smty'] = sf.data['smty'] * 1e10
    # extract some mass data (smty0=gas, smty4=star)
    sf.data['smty0'] = sf.data['smty'][:,0]
    sf.data['smty4'] = sf.data['smty'][:,4]
    # V-band magnitude
    sf.data['ssphv'] = sf.data['ssph'][:,2]
    return sf.data

def mergeSimData(sims):
    mergedDic = {}
    assert len(sims)!=0
    keys = sims[0]
    for key in keys:
        mergedDic[key] = np.concatenate([sim[key] for sim in sims])
    return mergedDic

# convert simulation quantity to its acronym if necessary
def checkSimVarName(name):
    if name in sim_var_acronyms.keys():
        name = sim_var_acronyms[name]
    return name

# plot relationships for both sim and obs
def plot2(sx, sy, ox, oy):
    sx_data = sim[checkSimVarName(sx)]
    sy_data = sim[checkSimVarName(sy)]
    ox_data, oy_data = obs[ox], obs[oy]
    
    plt.subplots(1,1)
    semilogy(sx_data, sy_data,'ro',alpha=0.5,label='sim')
#     xlabel(sim_var_units[sx])
#     ylabel(sim_var_units[sy])
#     title(sy + ' vs ' + sx)
    
    semilogy(ox_data, oy_data, 'bo',alpha=0.7,label='obs')
    legend(loc=0,numpoints=1)
    xlabel(obs_var_units[ox])
    ylabel(obs_var_units[oy])
    title(oy + ' vs ' + ox)
```

```python
"""
- Plot replationships between mass and distance-to-host of galaxies for simulation data and observation data.
- Use 2000kpc cutoff for all data.
- Select only galaxies with stars in simulation data.
- Use 5e6 stellar mass cutoff for the mass of stars of simulation data.

Four types of data:
blue dot: observation data, galaxies with gas, mass of HI vs. distance to the host
blue cross: observation data, galaxies without gas, HI limit vs. distance to the host
red dot: simulation data, galaxies with gas, mass of HI vs. distance to the host
red cross: simulation data, galaxies without gas, mass of stars vs. distance to the host
"""

def MHI_dist_graph(sim_nums):
    sims = [getSimData(sim_num) for sim_num in sim_nums]
    sim = mergeSimData(sims)
        
    # get raw data
    sx_data = sim[checkSimVarName('SubhaloDis')]
    sy_data = sim[checkSimVarName('SubhaloMassTypeGas')]
    star_mass = sim[checkSimVarName('SubhaloMassTypeStar')]
    ox_data = obs['min_dist_mw_or_m31']
    oy_data = obs['MHI_computed']
    
    # select simulation data with stars
    stellar_mass_cutoff = 5e5
    has_star_inds = star_mass > stellar_mass_cutoff
    sx_data = sx_data[has_star_inds]
    sy_data = sy_data[has_star_inds]
    star_mass = star_mass[has_star_inds]
    
    # distance cutoff
    dist_cutoff = 2000 #kpc
    sim_inds = sx_data <= dist_cutoff
    sx_data = sx_data[sim_inds]
    sy_data = sy_data[sim_inds]
    star_mass = star_mass[sim_inds]
    obs_inds = ox_data <= dist_cutoff
    ox_data = ox_data[obs_inds]
    oy_data = oy_data[obs_inds]

    fig, ax1 = plt.subplots(1,1)
    fig.set_size_inches(16.5, 10.5)

    # plot simulation data, with gas
    ax1.semilogy(sx_data, sy_data,'ro',alpha=0.7,label='sim: with gas')
    
    # plot simulation data, without gas
    no_gas_inds = sy_data == 0
    ax1.semilogy(sx_data[no_gas_inds], star_mass[no_gas_inds],'r+',alpha=0.7,label='sim: stellar mass limit', markersize=10)

    # plot observation data, with gas
    detected_inds = obs['MHI_type']=='detected_mass'
    ax1.semilogy(ox_data[detected_inds], oy_data[detected_inds], 'bo',alpha=0.7,label='obs: detected')

    # plot observation data, without gas
    limit_inds = obs['MHI_type']!='detected_mass'
    ax1.semilogy(ox_data[limit_inds], oy_data[limit_inds], 'b+',alpha=0.7,label='obs: limit', markersize=10)

    ax1.legend(loc=2,numpoints=1)
    ax1.set_xlabel('distance (kpc)')
    ax1.set_ylabel('mass (Msun)')
    
    if len(sim_nums) == 0:
        sim_num = str(sim_nums[0])
    else:
        sim_num = "%d-%d" % (sim_nums[0], sim_nums[-1])

    title('Mass of HI vs Distance to Host Galaxy (simulation %s)' % sim_num)
    plt.savefig("M_HI-vs-distance(sim %s)" % sim_num)
```

```python
"""
- Plot replationships between fraction of gas-poor galaxies vs. distance-to-host for simulation data and observation data.
- Use 2000kpc cutoff for all data.
- Select only galaxies with stars in simulation data.
- Use 5e6 stellar mass cutoff for the mass of stars of simulation data.
"""

def fraction_gas_poor_graph(sim_nums):
    sims = [getSimData(sim_num) for sim_num in sim_nums]
    sim = mergeSimData(sims)
        
    # get raw data
    sx_data = sim[checkSimVarName('SubhaloDis')]
    sy_data = sim[checkSimVarName('SubhaloMassTypeGas')]
    star_mass = sim[checkSimVarName('SubhaloMassTypeStar')]
    ox_data = obs['min_dist_mw_or_m31']
    oy_data = obs['MHI_computed']
    
    # select simulation data with stars
    stellar_mass_cutoff = 5e5
    has_star_inds = star_mass > stellar_mass_cutoff
    sx_data = sx_data[has_star_inds]
    sy_data = sy_data[has_star_inds]
    star_mass = star_mass[has_star_inds]
    
    # distance cutoff
    dist_cutoff = 2000 #kpc
    sim_inds = sx_data <= dist_cutoff
    sx_data = sx_data[sim_inds]
    sy_data = sy_data[sim_inds]
    star_mass = star_mass[sim_inds]
    obs_inds = ox_data <= dist_cutoff
    ox_data = ox_data[obs_inds]
    oy_data = oy_data[obs_inds]

    fig, ax1 = plt.subplots(1,1)
    fig.set_size_inches(10.5, 8.5)

    bins = np.arange(0,2001,250)
    
    # all sim data
    simNoGasBins, _ = np.histogram(sx_data[sy_data == 0], bins)
    simBins, _ = np.histogram(sx_data, bins)
    
    # group1: sim data 5e5 - 1e7
    group1 = np.where(np.logical_and(star_mass>=5e5, star_mass<1e7))
    g1NoGasBins, _ = np.histogram(sx_data[group1][sy_data[group1] == 0], bins)
    g1Bins, _ = np.histogram(sx_data, bins)
    
    # group2: sim data 1e7 - 1e8
    group2 = np.where(np.logical_and(star_mass>=1e7, star_mass<1e8))
    g2NoGasBins, _ = np.histogram(sx_data[group2][sy_data[group2] == 0], bins)
    g2Bins, _ = np.histogram(sx_data, bins)
    
     # group3: sim data > 1e8
    group3 = np.where(star_mass>=1e8)
    g3NoGasBins, _ = np.histogram(sx_data[group3][sy_data[group3] == 0], bins)
    g3Bins, _ = np.histogram(sx_data, bins)
    
    # obs data
    obsNoGasBins, _ = np.histogram(ox_data[obs['MHI_type']!='detected_mass'], bins)
    obsBins, _ = np.histogram(ox_data, bins)
        
    bin_centers = 0.5*(bins[1:]+bins[:-1])
    ax1.plot(bin_centers,simNoGasBins/simBins, label = 'simulation', marker='v', markersize='10')
    ax1.plot(bin_centers,g1NoGasBins/g1Bins, label = 'simulation: 5e5-1e7', marker='v', markersize='10')
    ax1.plot(bin_centers,g2NoGasBins/g2Bins, label = 'simulation: 1e7-1e8', marker='v', markersize='10')
    ax1.plot(bin_centers,g3NoGasBins/g3Bins, label = 'simulation: >1e8', marker='v', markersize='10')
    ax1.plot(bin_centers,obsNoGasBins/obsBins, label = 'observation', marker='o', markersize='10')
    
    ax1.set_ylim(0, 1)
    ax1.legend(loc=0,numpoints=1)
    ax1.set_xlabel('distance (kpc)')
    
    if len(sim_nums) == 1:
        sim_num = str(sim_nums[0])
    else:
        sim_num = "%d-%d" % (sim_nums[0], sim_nums[-1])

    title('Fraction of Gass Poor Galaxies vs Distance to Host Galaxy (simulation %s)' % sim_num)
    plt.savefig("fraction_gas_poor-vs-distance(sim only %s)" % sim_num)
```
